```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import *
```

```python
df_train = pd.read_csv("sales_train.csv")
df_train['date']=df_train['date'].apply(lambda x:datetime.datetime.strptime(x,'%d.%m.%Y'))
df_train['month_year'] = df_train['date'].dt.to_period('M')
df_train
```

```python
shop_id
```

```python
def shop(n):
    shop_25 = df_train[df_train['shop_id'] == n]

    shop_25["revenue"] = shop_25["item_price"] * shop_25["item_cnt_day"]

    grouped_df = shop_25.groupby(['month_year'])['month_year','revenue'].agg({'sum'})
    grouped_df = grouped_df.reset_index()
    grouped_df.set_index(['month_year'], inplace=True)
    grouped_df.rename(columns = {'item_cnt_day':'item_cnt_month'}, inplace = True)
    # grouped_df = grouped_df.to_timestamp()
    plt.figure(figsize=(10, 8))
    plt.title("shop " + str(n))
    plt.plot(grouped_df.to_timestamp(), label='Actual data')
    #plt.fill_between(period_index.to_timestamp(), confint_df[0], confint_df[1],color='grey',alpha=.2, label='Confidence Intervals Area')
    plt.legend()
    plt.show()
```

```python
for x in shop_id:
    shop(x)
```

```python
# check stationary of revenue
    
df_train["revenue"] = df_train["item_price"] * df_train["item_cnt_day"]
grouped_df = df_train.groupby(['month_year'])['month_year','revenue'].agg({'sum'})
grouped_df = grouped_df.reset_index()
grouped_df.set_index(['month_year'], inplace=True)
grouped_df.rename(columns = {'item_cnt_day':'item_cnt_month'}, inplace = True)
# grouped_df = grouped_df.to_timestamp()
plt.figure(figsize=(10, 8))
plt.title("REVENUE")
plt.plot(grouped_df.to_timestamp(), label='Actual data')
#plt.fill_between(period_index.to_timestamp(), confint_df[0], confint_df[1],color='grey',alpha=.2, label='Confidence Intervals Area')
plt.legend()
plt.show()
    
t_stat, p_value, critical_values, _, _ = zivot_andrews(grouped_df["revenue"].values)
print(f'Zivot-Andrews Statistic: {t_stat:.2f}')
for key, value in critical_values.items():
    print('Critial Values:')
    print(f'   {key}, {value:.2f}')

print(f'\np-value: {p_value:.2f}')
print("Non-Stationary") if p_value > 0.05 else print("Stationary")
```

```python
#plot and check stationary of item_cnt
    
df_train["revenue"] = df_train["item_price"] * df_train["item_cnt_day"]
grouped_df = df_train.groupby(['month_year'])['month_year','item_cnt_day'].agg({'sum'})
grouped_df = grouped_df.reset_index()
grouped_df.set_index(['month_year'], inplace=True)
grouped_df.rename(columns = {'item_cnt_day':'item_cnt_month'}, inplace = True)
# grouped_df = grouped_df.to_timestamp()
plt.figure(figsize=(10, 8))
plt.title("ITEM_CNT")
plt.plot(grouped_df.to_timestamp(), label='Actual data')
#plt.fill_between(period_index.to_timestamp(), confint_df[0], confint_df[1],color='grey',alpha=.2, label='Confidence Intervals Area')
plt.legend()
plt.show()
    
#zivot_andrews root test
t_stat, p_value, critical_values, a, b = zivot_andrews(grouped_df["item_cnt_month"].values)
print(f'Zivot-Andrews Statistic: {t_stat:.2f}')
for key, value in critical_values.items():
    print('Critial Values:')
    print(f'   {key}, {value:.2f}')

print(f'\np-value: {p_value:.2f}')
print("Non-Stationary") if p_value > 0.05 else print("Stationary")
```
